#### Librerias

```python
import numpy as np
from keras.models import Sequential
from keras.layers.core import Dense
```

#### Se cargan las entradas para la compuerta XOR.

```python
datos_entrenamiento = np.array([[0,0],[0,1],[1,0],[1,1]], "float32")

datos_entrenamiento
```

#### Se cargan las salidas para la compuerta XOR. 

```python
datos_salida = np.array([[0],[1],[1],[0]], "float32")
```

#### Se programa la Red Neuronal

```python
model = Sequential()
model.add(Dense(8, input_dim=2, activation='relu'))
model.add(Dense(1, activation='sigmoid'))

model.compile(loss='mean_squared_error', optimizer='adam', metrics=['binary_accuracy'])

model.fit(datos_entrenamiento, datos_salida, epochs=400)
```

#### Se evaluá el modelo

```python
scores = model.evaluate(datos_entrenamiento, datos_salida)

print(model.metrics_names[1], scores[1]*100)
print(model.predict(datos_entrenamiento).round())
```
