#### Librerias

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


# Libreria para la normalización de datos
from sklearn.preprocessing import normalize
# Libreria para el dendograma
from scipy.cluster.hierarchy import dendrogram, linkage
# Libreria para el cluster Aglomerativo
from sklearn.cluster import AgglomerativeClustering
```

#### Cargue de los datos

```python
data = pd.read_csv('data/wholesale_customer.csv')
data.head()
```

#### Normalización de datos

```python
##### INSERTE SU CÓDIGO AQUÍ #####
# Se normalizan los datos
datos_normalizados = normalize(data)
datos_normalizados = pd.DataFrame(datos_normalizados, columns=data.columns)
datos_normalizados.head(10)


##### FIN CÓDIGO INSERTADO #####
```

#### Dendograma

```python
##### INSERTE SU CÓDIGO AQUÍ #####
# Gráfica Dendograma con método 'ward'
import scipy.cluster.hierarchy as shc

plt.figure(figsize=(10, 7))
plt.title("Dendograma")
dendograma = shc.dendrogram(shc.linkage(datos_normalizados, method='ward')) 
##### FIN CÓDIGO INSERTADO #####
```

#### Umbral

```python
##### INSERTE SU CÓDIGO AQUÍ #####
# Dendograma con Umbral
plt.figure(figsize=(10, 7))
plt.title("Dendrograma")
dendograma = shc.dendrogram(shc.linkage(datos_normalizados, method='ward'))
plt.axhline(y=6, color='r', linestyle='--')

##### FIN CÓDIGO INSERTADO #####
```

#### Clusters Aglomerativo

```python
##### INSERTE SU CÓDIGO AQUÍ #####
# Cluster Aglomerativo con: 2 clusters, afinidad euclidiana y enlace ward
ac = AgglomerativeClustering(n_clusters=2, metric='euclidean', linkage='ward')

ac.fit(datos_normalizados)

labels = ac.labels_

labels

##### FIN CÓDIGO INSERTADO #####
```

#### Visualización de los grupos

```python
##### INSERTE SU CÓDIGO AQUÍ #####
# Visualización de los grupos: 'Milk' eje X y 'Grocery' eje Y
plt.figure(figsize=(10, 7))
plt.scatter(datos_normalizados['Milk'], datos_normalizados['Grocery'], c=ac.labels_)
##### FIN CÓDIGO INSERTADO #####

```
